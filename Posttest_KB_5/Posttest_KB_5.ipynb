Handling Outliers

```python
# Penanganan Outlier dengan Metode IQR (Interquartile Range)

from scipy import stats

# Sebelum menangani outlier (visualisasi awal)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.tight_layout()
sns.boxplot(df['salary'], ax=axes[0])
axes[0].set_title("Sebelum Penanganan Outlier")

# Menangani outlier dengan IQR
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

# Hapus outlier di luar 1.5 * IQR
df['salary'] = df['salary'][~((df['salary'] < (Q1 - 1.5 * IQR)) | (df['salary'] > (Q3 + 1.5 * IQR)))]

# Setelah menangani outlier (visualisasi setelah)
sns.boxplot(df['salary'], ax=axes[1])
axes[1].set_title("Setelah Penanganan Outlier")
plt.show()
```

Support Vector Classifier (dari SVM)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score

# Load dataset dari file CSV
df = pd.read_csv('jobs_in_data.csv')

# Pilih 2 fitur untuk klasifikasi biner
X = df[['experience_level', 'employment_type']].copy()  # Gunakan .copy() untuk menghindari peringatan
y = np.where(df['experience_level'] == 'Entry-level', 1, -1)  # Ubah sesuai kebutuhan

# Encoding fitur kategorikal menggunakan .loc untuk menghindari peringatan
le_X = LabelEncoder()
X['experience_level'] = le_X.fit_transform(X['experience_level'])
X['employment_type'] = le_X.fit_transform(X['employment_type'])

# Bagi data menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fungsi untuk melatih dan mengevaluasi model
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train.values, y_train)  # Menggunakan .values
    y_pred = model.predict(X_test.values)  # Menggunakan .values
    print(f"Accuracy of {model.__class__.__name__}: {accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))
    return y_pred

# SVM
svm = SVC(kernel="linear")
print("Support Vector Machine")
y_pred_svm = evaluate_model(svm, X_train, y_train, X_test, y_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
print("\nRandom Forest")
y_pred_rf = evaluate_model(rf, X_train, y_train, X_test, y_test)

# K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)
print("\nK-Nearest Neighbors")
y_pred_knn = evaluate_model(knn, X_train, y_train, X_test, y_test)

# Naive Bayes
nb = GaussianNB()
print("\nNaive Bayes")
y_pred_nb = evaluate_model(nb, X_train, y_train, X_test, y_test)

# Visualisasi SVM Decision Boundary
plt.figure(figsize=(14, 10))

# SVM Visualization
plt.subplot(2, 2, 1)
plt.scatter(X_train.iloc[y_train == 1, 0], X_train.iloc[y_train == 1, 1],
            s=40, label='Entry-level', color='green', marker='o')
plt.scatter(X_train.iloc[y_train == -1, 0], X_train.iloc[y_train == -1, 1],
            s=40, label='Non Entry-level', color='red', marker='o')

# Tampilkan support vectors
support_vectors = svm.support_vectors_
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, color='cyan', marker='+', label='Support Vectors')

# Buat garis decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1]), np.linspace(ylim[0], ylim[1]))
Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary dan margins
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.xlabel('Experience Level (Encoded)')
plt.ylabel('Employment Type (Encoded)')
plt.legend(loc='upper right')
plt.title('SVM dengan Decision Boundary')

# Random Forest Visualization (titik prediksi)
plt.subplot(2, 2, 2)
plt.scatter(X_test.iloc[y_test == 1, 0], X_test.iloc[y_test == 1, 1],
            s=40, label='Entry-level', color='green', marker='o')
plt.scatter(X_test.iloc[y_test == -1, 0], X_test.iloc[y_test == -1, 1],
            s=40, label='Non Entry-level', color='red', marker='o')
plt.scatter(X_test.iloc[y_pred_rf == 1, 0], X_test.iloc[y_pred_rf == 1, 1],
            s=100, edgecolor='blue', facecolor='none', label='Predicted Entry-level (RF)', marker='o')
plt.scatter(X_test.iloc[y_pred_rf == -1, 0], X_test.iloc[y_pred_rf == -1, 1],
            s=100, edgecolor='orange', facecolor='none', label='Predicted Non Entry-level (RF)', marker='o')
plt.xlabel('Experience Level (Encoded)')
plt.ylabel('Employment Type (Encoded)')
plt.legend(loc='upper right')
plt.title('Random Forest Prediksi')

# K-Nearest Neighbors Visualization (titik prediksi)
plt.subplot(2, 2, 3)
plt.scatter(X_test.iloc[y_test == 1, 0], X_test.iloc[y_test == 1, 1],
            s=40, label='Entry-level', color='green', marker='o')
plt.scatter(X_test.iloc[y_test == -1, 0], X_test.iloc[y_test == -1, 1],
            s=40, label='Non Entry-level', color='red', marker='o')
plt.scatter(X_test.iloc[y_pred_knn == 1, 0], X_test.iloc[y_pred_knn == 1, 1],
            s=100, edgecolor='blue', facecolor='none', label='Predicted Entry-level (KNN)', marker='o')
plt.scatter(X_test.iloc[y_pred_knn == -1, 0], X_test.iloc[y_pred_knn == -1, 1],
            s=100, edgecolor='orange', facecolor='none', label='Predicted Non Entry-level (KNN)', marker='o')
plt.xlabel('Experience Level (Encoded)')
plt.ylabel('Employment Type (Encoded)')
plt.legend(loc='upper right')
plt.title('K-Nearest Neighbors Prediksi')

# Naive Bayes Visualization (titik prediksi)
plt.subplot(2, 2, 4)
plt.scatter(X_test.iloc[y_test == 1, 0], X_test.iloc[y_test == 1, 1],
            s=40, label='Entry-level', color='green', marker='o')
plt.scatter(X_test.iloc[y_test == -1, 0], X_test.iloc[y_test == -1, 1],
            s=40, label='Non Entry-level', color='red', marker='o')
plt.scatter(X_test.iloc[y_pred_nb == 1, 0], X_test.iloc[y_pred_nb == 1, 1],
            s=100, edgecolor='blue', facecolor='none', label='Predicted Entry-level (NB)', marker='o')
plt.scatter(X_test.iloc[y_pred_nb == -1, 0], X_test.iloc[y_pred_nb == -1, 1],
            s=100, edgecolor='orange', facecolor='none', label='Predicted Non Entry-level (NB)', marker='o')
plt.xlabel('Experience Level (Encoded)')
plt.ylabel('Employment Type (Encoded)')
plt.legend(loc='upper right')
plt.title('Naive Bayes Prediksi')

plt.tight_layout()
plt.show()
```

Penjelasan Metode
1. Support Vector Machine (SVM):
SVM adalah algoritma klasifikasi yang mencari hyperplane terbaik yang memisahkan kelas-kelas di ruang fitur. SVM sangat efektif dalam kasus di mana jumlah fitur lebih besar dibandingkan jumlah sampel.
Kelebihan: Kinerja yang baik dalam dataset yang kompleks dan dapat digunakan untuk klasifikasi linier dan non-linier (menggunakan kernel).
2. Random Forest:
Random Forest adalah metode ensemble yang menggunakan banyak pohon keputusan untuk memberikan hasil akhir. Setiap pohon memberikan suara untuk kelas dan kelas dengan suara terbanyak menang.
Kelebihan: Tahan terhadap overfitting dan memberikan estimasi yang baik dari fitur penting. Sangat efektif untuk dataset besar.
3. K-Nearest Neighbors (KNN):
KNN adalah algoritma yang mengklasifikasikan data berdasarkan jarak ke titik-titik terdekat dalam ruang fitur. Ketika memprediksi kelas suatu titik, KNN mencari k titik terdekat dan mengambil suara mayoritas dari titik-titik tersebut.
Kelebihan: Sederhana dan efektif, serta mudah diimplementasikan. Tidak memerlukan pelatihan, tetapi memerlukan penyimpanan data pelatihan.
4. Naive Bayes:
Naive Bayes adalah classifier probabilistik berdasarkan Teorema Bayes dengan asumsi independensi antar fitur. Ini sangat cepat dan efektif untuk klasifikasi teks.
Kelebihan: Sangat efisien dalam hal waktu dan memori, bekerja baik dengan data yang besar, dan cocok untuk klasifikasi teks.

Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

# Load dataset dari file CSV
df = pd.read_csv('jobs_in_data.csv')

# Misalkan kita ingin memprediksi 'salary' berdasarkan 'experience_level' yang sudah di-encode
# Gantilah 'salary' dengan kolom target numerik yang ada dalam dataset Anda.
X = df[['experience_level', 'employment_type']].copy()  # Fitur
y = df['salary']  # Target numerik, ganti dengan kolom yang sesuai

# Encoding fitur kategorikal menggunakan LabelEncoder
le_X = LabelEncoder()
X['experience_level'] = le_X.fit_transform(X['experience_level'])
X['employment_type'] = le_X.fit_transform(X['employment_type'])

# Bagi data menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Regresi Linear
# Buat objek Regresi Linear dari Library Scikit-Learn
linear_reg = LinearRegression()
# Latih model dengan variabel X dan Y dari Train Set
linear_reg.fit(X_train, y_train)
# Prediksi variabel Y dengan menggunakan variabel X dari Test Set
y_pred_linear = linear_reg.predict(X_test)

# Visualisasi hasil prediksi Regresi Linear
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.scatter(y_test, y_pred_linear, color="black")
plt.plot(y_test, y_test, color="blue", linewidth=3)  # Garis ideal
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Regresi Linear: Predicted vs Actual')

# 2. Support Vector Regressor
svr = SVR(kernel='linear')
# Latih model SVR
svr.fit(X_train, y_train)
# Prediksi menggunakan SVR
y_pred_svr = svr.predict(X_test)

# Visualisasi hasil prediksi SVR
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.scatter(y_test, y_pred_svr, color="red")
plt.plot(y_test, y_test, color="blue", linewidth=3)  # Garis ideal
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Support Vector Regressor: Predicted vs Actual')

plt.tight_layout()
plt.show()
```

Penjelasan Metode
1. Regresi Linear:
Regresi linear adalah metode statistik yang digunakan untuk memodelkan hubungan antara variabel independen (fitur) dan variabel dependen (target). Tujuannya adalah menemukan garis terbaik yang dapat memprediksi nilai target berdasarkan fitur.
Kelebihan: Mudah diinterpretasikan, cepat dalam pelatihan, dan bekerja dengan baik ketika hubungan antara variabel bersifat linier.
Kekurangan: Sensitif terhadap outlier dan tidak bekerja dengan baik jika hubungan antara variabel adalah non-linier.
2. Support Vector Regressor (SVR):
SVR adalah versi regresi dari Support Vector Machine (SVM). SVR mencoba menemukan hyperplane yang paling sesuai untuk memprediksi nilai target sambil meminimalkan kesalahan.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load dataset dari file CSV
df = pd.read_csv('jobs_in_data.csv')

# Pilih 2 fitur untuk klasifikasi biner
X = df[['experience_level', 'employment_type']].copy()
y = np.where(df['experience_level'] == 'Entry-level', 1, -1)  # Ubah sesuai kebutuhan

# Encoding fitur kategorikal
le_X = LabelEncoder()
X['experience_level'] = le_X.fit_transform(X['experience_level'])
X['employment_type'] = le_X.fit_transform(X['employment_type'])

# Bagi data menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fungsi untuk melatih dan mengevaluasi model
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train.values, y_train)  # Menggunakan .values
    y_pred = model.predict(X_test.values)  # Menggunakan .values

    # Evaluasi model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of {model.__class__.__name__}: {accuracy}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non Entry-level', 'Entry-level'], yticklabels=['Non Entry-level', 'Entry-level'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - {model.__class__.__name__}')
    plt.show()

    # Classification Report
    report = classification_report(y_test, y_pred, target_names=['Non Entry-level', 'Entry-level'])
    print(report)

    return y_pred

# SVM
svm = SVC(kernel="linear")
print("Support Vector Machine")
y_pred_svm = evaluate_model(svm, X_train, y_train, X_test, y_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
print("\nRandom Forest")
y_pred_rf = evaluate_model(rf, X_train, y_train, X_test, y_test)

# K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)
print("\nK-Nearest Neighbors")
y_pred_knn = evaluate_model(knn, X_train, y_train, X_test, y_test)

# Naive Bayes
nb = GaussianNB()
print("\nNaive Bayes")
y_pred_nb = evaluate_model(nb, X_train, y_train, X_test, y_test)
```

Regresi: MAE, MSE, RMSE, dan R2-Score

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR  # Gunakan SVR untuk regresi
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Memuat dataset Iris dari Seaborn
iris = sns.load_dataset('iris')

# Memilih 2 fitur untuk regresi (petal length dan petal width) dan menjadikan 'sepal length' sebagai target
X = iris[['petal_length', 'petal_width']]
y = iris['sepal_length']  # Target yang akan diprediksi

# Bagi data menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat objek regresi SVM
svm_reg = SVR(kernel="linear")  # Gunakan kernel linear untuk regresi
svm_reg.fit(X_train, y_train)  # Latih model dengan data train

# Lakukan prediksi pada data test
y_pred = svm_reg.predict(X_test)

# Menghitung metrik evaluasi
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Menampilkan hasil evaluasi
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-Squared Score:", r2)
```

Penjelasan Kode:
1. Import Library: Kode ini menggunakan library numpy, pandas, dan seaborn untuk memanipulasi data, serta SVR dari sklearn.svm untuk melakukan regresi.
2.Load Dataset: Dataset Iris diambil dari Seaborn, di mana kita memilih dua fitur (petal_length dan petal_width) untuk digunakan sebagai fitur input dan sepal_length sebagai target yang akan diprediksi.
3. Train-Test Split: Data dibagi menjadi set pelatihan (train set) dan set pengujian (test set) dengan proporsi 70:30.
4. Model Regresi SVM: Membuat objek SVR (Support Vector Regression) dengan kernel linear. Model dilatih menggunakan data pelatihan.
5. Prediksi: Model melakukan prediksi terhadap data pengujian.
6. Menghitung Metrik Evaluasi:
* Mean Absolute Error (MAE): Mengukur rata-rata kesalahan absolut antara prediksi dan nilai aktual.
* Mean Squared Error (MSE): Mengukur rata-rata kesalahan kuadrat antara prediksi dan nilai aktual. MSE memberikan bobot lebih pada kesalahan yang lebih besar.
* Root Mean Squared Error (RMSE): Akar kuadrat dari MSE, yang mengembalikan kesalahan ke unit asli dari target.
R-Squared Score (R²): Mengukur proporsi variasi dalam data target yang dapat dijelaskan oleh model. Nilai R² berkisar antara 0 dan 1, di mana nilai yang lebih tinggi menunjukkan model yang lebih baik.
